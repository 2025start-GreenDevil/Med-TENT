### 결측치 제거된 파일 읽어오기 (hos73 vs hos175)

```python
hos73 = pd.read_csv('/content/drive/MyDrive/greenDevil/removed_hospital73.csv')
hos73
```

```python
hos175 = pd.read_csv('/content/drive/MyDrive/greenDevil/removed_hospital175.csv')
hos175
```

### 전체 입퇴원 기간(totaloffset) 추가

```python
hos73['totaloffset'] = hos73['hospitaldischargeoffset'] + hos73['hospitaladmitoffset'].abs()
hos175['totaloffset'] = hos175['hospitaldischargeoffset'] + hos175['hospitaladmitoffset'].abs()
```

```python
hos73
```

```python
hos175
```

### totaloffset이 1년 넘어가는지 체크
1년=525600 offset

```python
hos73[hos73['totaloffset']>525600]
```

```python
hos175[hos175['totaloffset']>525600]
```

```python
# hos73는 offset 최대 8개월
hos73[['totaloffset']].sort_values(by='totaloffset', ascending=False)
```

```python
# hos175는 offset 최대 1개월
hos175[['totaloffset']].sort_values(by='totaloffset', ascending=False)
```

### 입원 년도(admit year) 추가
totaloffset이 두 병원 모두 1년이 안넘어가므로 퇴원 년도와 동일

```python
hos73['hospitaladmityear'] = hos73['hospitaldischargeyear']
hos175['hospitaladmityear'] = hos175['hospitaldischargeyear']
```

```python
hos175
```
