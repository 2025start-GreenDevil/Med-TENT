```python
def clean_icd_code(code):
    if pd.isnull(code):
        return code  # Leave NaNs unchanged

    # comma 기준으로 진단 코드 분리
    parts = [p.strip() for p in str(code).split(',') if p.strip() != '']

    # If any part is an ICD-9 code (usually all digits), keep the first such code
    for part in parts:
        if part.replace('.', '').isdigit():
            return part  # ICD-9
    # ICD-9 진단 코드가 없다면, 첫 번째 ICD-10 진단 코드 반환
    return parts[0] if parts else None

# icd9code 칼럼에 함수 적용
df['icd9code'] = df['icd9code'].apply(clean_icd_code)
```
